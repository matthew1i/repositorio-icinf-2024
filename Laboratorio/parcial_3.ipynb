```python
lista2 = lista.dropna()
lista2.isnull().any()
```

```python
listap = lista2["Win"].mean
listap  
```

```python
listag = lista.groupby("Team")["Goals For"].sum()
top = listag.nlargest(10)
plt.figure(figsize=(10, 6))
top.plot(kind='bar', color='skyblue')
plt.title('Cantidad de Goles Anotados por el top 10 de Pases')
plt.xlabel('Pas')
plt.ylabel('Goles')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
